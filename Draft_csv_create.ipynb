```python
import csv
import re
import os           # This is for os.listdir
import os.path      # This is for the other dir stuff.
import string       # Maybe for directory name cycling etc.
import time # For timing how long it takes.
import subprocess

line_re = re.compile(r'(\s\W(\d*)\W\s\W\s)([\W])([\d]),(\s.*\s?\d*\s?\d*\s?\d*\s.*\s?\d*\s?\d*\s?\d*),(\s.*\d*),(\s*\d*),(\s*\d*),(\s.*\d*),(\s.*[A-Z].*),(\s.*[A-Z].*),(\s.*[A-Z].*),(\s.*[a-z].*),(\s.*\d*)([\W\W])$')
   
def parse_and_write(file):
    for line in file:                           # Oh the first "line" is a hard return???
        #print 'A line is: ', line
        data = line_re.match(line)
        if data is not(None):
            line_no = data.group(2)
            timestamp = data.group(5)
            seq_no = data.group(6)
            avatar = data.group(7)
            #Guild = data.group(8)
            Level = data.group(9)
            Race = data.group(10)
            Class = data.group(11)
            Zone = data.group(12)

            if data.group(8) is not(' '):
                Guild = data.group(8)
            else:
                Guild = '-1'   # Note there are some missing values, i.e. errant -1.
            #print timestamp    # This is so you can keep track of where it is. Max Jan 2009 IIRC.

            new_line = line_no + ',' + seq_no + ',' + avatar + ',' + Guild + ',' + Level + ',' + Race + ',' + Class + ',' + Zone +  ',' + timestamp + '\n'
            #print new_line
            output_file.write(new_line)
                
        else:
            print "Didn't match the regex."

def main():
    global output_file
    write_data_filename = 'wowah.csv'
    output_file = open(write_data_filename, 'w')    # 'a' is very important, it appends the new data to the big file. 
    fieldnames = ('line_no, seq_no, avatar, Guild, Level, Race, Class,Zone,timestamp\n')
    output_file.write(fieldnames)
    cmd = 'hdfs dfs -find {} -name *.txt'.format('/user/deepannrr3037/Checks').split()
    files = subprocess.check_output(cmd).strip().split('\n')
    for path in files:
        cat = subprocess.Popen(["hadoop", "fs", "-cat", path], stdout=subprocess.PIPE)
        parse_and_write(cat.stdout)
            
main()
```
